#### Plot

```python
import numpy as np
puntos_x = np.array([0, 4, 8, 16])
puntos_y = np.array([6, 12, 24, 48])
plt.plot(puntos_x, puntos_y)
plt.show()
```

```python
x1= np.array([0, 1, 2, 3])
y1= np.array([3, 8, 1, 10])
x2= np.array ([0, 1, 2, 3])
y2= np.array([6, 2, 7, 11])

plt.plot(x1, y1, x2, y2)
plt.show()
```

#### Formato

Diagrama de puntos:

```python
puntos_y = np.array([6, 12, 24, 48])
plt.plot(puntos_y, 'o')
plt.show()
```

Argumentos:

<br>

Tana√±o: <br>
<br>
'o--k': punto con doble l√≠nea en color negro <br>

T√≠tulos<br>

```python
x=np.array([80, 85, 95, 100, 105, 110, 115, 120, 125])
y=np.array([240, 250, 270, 280, 290, 300, 310, 320, 330])
plt.plot(x, y)
plt.title("Estad√≠sticas deportivas")
plt.xlabel("Pulso Promedio")
plt.ylabel("Quema de calor√≠as")
plt.show()
```
